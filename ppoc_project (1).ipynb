# Infant Mortality Rate Analysis

```python
imr_total_data = pd.read_excel("/content/1-IMR.xlsx",sheet_name="Total",index_col=0,header = 3,skipfooter = 1)
imr_urban_data = pd.read_excel("/content/1-IMR.xlsx",sheet_name="Urban",index_col=0,header = 3,skipfooter = 1)
imr_rural_data = pd.read_excel("/content/1-IMR.xlsx",sheet_name="Rural",index_col=0,header = 3,skipfooter = 1)
imr_total_data.head()
```

```python
imr_total = imr_total_data.T
imr_urban = imr_urban_data.T
imr_rural = imr_rural_data.T
imr_total.head()
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import plotly.graph_objects as go

#extract color palette, the palette can be changed
pal = list(sns.color_palette(palette='viridis', n_colors=len(imr_total.columns)).as_hex())

fig = go.Figure()
for d,p in zip(imr_total.columns, pal):
    fig.add_trace(go.Scatter(y = imr_total[d],x=imr_total.index,
                             name = d,
                             line_color = p, 
                             fill=None))   
fig.show()
```

```python
plt.rcParams["figure.figsize"] = (8,5)
ax = plt.axes()
plt.plot(imr_total['India'],label="Total IMR")
plt.plot(imr_urban['India'],label="Urban IMR")
plt.plot(imr_rural['India'],label="Rural IMR")
ax.xaxis.set_major_locator(plt.MaxNLocator(5))
plt.legend()
plt.show()
```

**Q2.List some of the potential issues that could arise through the use of the data of your choosing?**

Ans: The dataset is not evenly ordered at some points urban, rural data is segregated and at some other verticals it is not. The time scale of data is different in different dataset making it difficult to portray on a single scale.
